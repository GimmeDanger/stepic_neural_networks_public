```python
import numpy as np

np.random.seed(42)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_prime(x):
    return sigmoid(x) * (1 - sigmoid(x))

def get_error_L(sums, y):
    """
    compute error on the last layer of network
    y - ndarray of shape (n, 1)
    sums - ndarray of shape (n, 1)
    """
    return np.reshape(((sigmoid(sums) - y) * sigmoid_prime(sums)).mean(0), (1, 1))

def get_error_l(deltas, sums, weights):
    """
    compute error on the previous layer of network
    deltas - ndarray of shape (n, n_{l+1})
    sums - ndarray of shape (n, n_l)
    weights - ndarray of shape (n_{l+1}, n_l)
    """
    return (np.dot(deltas, weights) * sigmoid_prime(sums)).mean(0)
```

```python
a_0 = np.array([[0], [1], [2]])
W_1 = np.array([[2, 2, 2], [2, 2, 2]])
```

```python
z_1 = np.dot(W_1, a_0)
a_1 = sigmoid(z_1)
W_2 = np.array([[1, 1]])
a_1
```

```python
z_2 = np.dot(W_2, a_1)
a_2 = sigmoid(z_2)
a_2
```

```python
y = np.array([[1]])
y
```

```python
delta_2 = get_error_L(z_2, y)
delta_2
```

```python
delta_1 = get_error_l(delta_2, z_1, W_2)
delta_1
```

```python
ans = 2 * delta_1
ans
```

```python
a_0 = np.array([[0], [1], [2]])
W_1 = np.array([[8, 7, 8], [10, 10, 9]])
W_2 = np.array([[10, 9]])

z_1 = np.dot(W_1, a_0)
a_1 = sigmoid(z_1)
z_2 = np.dot(W_2, a_1)
a_2 = sigmoid(z_2)

y = np.array([[1]])
delta_2 = get_error_L(z_2, y)
delta_1 = get_error_l(delta_2, z_1, W_2)

ans = 2 * delta_1
ans
```
